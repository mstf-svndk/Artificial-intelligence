```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline  
# grafikleri dosyanın içindede gösterir
```

```python
x=np.arange(20) #20 tane rastgele rakam üret
y=np.random.normal(10,1,20) 
# ortalaması 10 olan 1 standart sapma olan ve 20 tane olan sayılar üret
z=np.random.normal(10,2,20)
```

```python
fig=plt.figure() # bir figür içine 2 3 tane grafik atılabilir
ax=fig.add_subplot(111)#figür içindeki grafik sayısı biz 1 tane seçtik
ax.plot(x,y,color="g") # eksenlerin ismi
ax.set_title("Line chart") # grafiğin ismi
ax.set_xlabel("x")
ax.set_ylabel("y")
plt.show()
```

```python
fig=plt.figure() 
ax=fig.add_subplot(111)
ax.plot(x,y,label="Thing #1") 
ax.plot(x,z,label="Thing #2")
ax.legend()
plt.show()
```

```python
fig=plt.figure()
ax=fig.add_subplot(111)
ax.plot(x,y,marker="o",c="#FF0000") # renk için c şeklinde kısaltma oldu renk kodu ile de yazabiliriz
ax.set_ylabel("$$$ Money",fontsize=10)
ax.set_xlabel("Years",fontsize=20)
fig.savefig("figure.png",dpi=300)
plt.show()
```

```python
fig2=plt.figure(figsize=(15,5))
a=fig2.add_subplot(131) 
a.scatter(x,y)
a.set_title("Scatter plot")

b=fig2.add_subplot(132) # satır sayısı, sütun sayısı, grafik sayısı 132 nin anlamı
b.hist(y)
b.set_title("Histogram")

c=fig2.add_subplot(133)
c.boxplot([y,z])
c.set_title("Box plot")

fig2.suptitle("Visualization")
plt.show()
```
