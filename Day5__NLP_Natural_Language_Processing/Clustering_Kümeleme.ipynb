```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
df=pd.read_excel("cars.xls")
```

```python
df.sample(3)
```

```python
x=df[["Price","Mileage","Cylinder"]]
```

```python
x.head()
```

```python
model=KMeans(n_clusters=4) #veriyi 4 e böldük
```

```python
model=model.fit(x)
```

```python
pred=model.predict(x)
```

```python
pred
```

```python
x.head()
```

```python
silhouette_score(x,pred)
```

```python
clust=x.assign(Cluster=model.labels_)
model.cluster_centers_.astype("int")
model.labels_
```

```python
plt.figure(figsize=(10,6))
plt.scatter(x.Mileage,x.Price,c=pred)
plt.title("Kaç gruba ayırmalıyım")
plt.xlabel("Kaç KM")
plt.ylabel("Fiyat ne kadar")
plt.legend()
plt.colorbar()
plt.show()
# fiyat ve km ye göre gruplama yaptık x ekseni km yi gösterir
```

```python
allscore=[]
allclusters=[]
def clust(clusters):
    for no_of_cluster in np.arange(1,clusters):
        no_of_cluster+=1
        model=KMeans(n_clusters=no_of_cluster)
        pred=model.fit_predict(x)
        score=silhouette_score(x,pred)
        print(f'Number of cluster {no_of_cluster}, silhoutte {score}')
        allscore.append(score)
        allclusters.append(no_of_cluster)
clust(10)
# burada kaç gruba ayıracağımızı bulmak için yaptık 4. gruptan itibaren değişim artık en az o zaman 4 e ayırdık
```

```python
plt.plot(allclusters,allscore)
plt.xlabel('Clusters()')
plt.ylabel('Silhoutte_Score')
plt.title('Silhouette Method')
plt.xlim(0,15)
plt.show()
```

```python
allscore=[]
allclusters=[]
sum_of_squared_distances = []
def clust(clusters):
    for no_of_cluster in np.arange(1,clusters):
        
        no_of_cluster+=1
        model=KMeans(n_clusters=no_of_cluster)
        #model-model fit(x) 
        pred=model.fit_predict(x)
        sum_of_squared_distances.append(model.inertia_)
        score=silhouette_score(x,pred)
        print ("Number of cluster {}, silhouette {}".format(no_of_cluster,score))
        allscore.append(score)
        allclusters.append(no_of_cluster)
clust(9)
```

```python
plt.plot(allclusters,sum_of_squared_distances,marker='x')
plt.xlabel('k')
plt.ylabel('Distortion')
plt.title('The Elbow Method showing optimal K')
plt.show()
# dirseğin kırıldığı noktaya bakarak kaç grup yapalım ona karar verebiliriz
```
