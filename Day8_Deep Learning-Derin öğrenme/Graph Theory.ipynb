```python
import pandas as pd
```

```python
df=pd.read_csv("graphdata.csv")
```

```python
df.head()
```

```python
import networkx as nx # graph teori için bunu içeri aldık
df=nx.from_pandas_edgelist(df,source="Origin", target="Dest",edge_attr=True)
```

```python
df.nodes() # havaalanları
```

```python
df.edges() # havaalanı arasındaki yollar
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.figure(figsize=(12,8))
nx.draw_networkx(df, with_labels=True)
```
