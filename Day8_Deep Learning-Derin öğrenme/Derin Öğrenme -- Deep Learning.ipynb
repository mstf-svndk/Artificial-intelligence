```python
df.columns=["preg", "plas", "pres", "skin", "test", "mass", "pedi", "age", "class"]
#sütun isimlerini değiştirdik
```

```python
df.head()
```


* yukarıdaki başlıkların anlamı

```python
x=df.iloc[:,0:8]
```

```python
y=df.iloc[:,8]
```

```python
model=Sequential()
model.add(Dense(12,activation="relu"))
model.add(Dense(8,activation="relu"))
model.add(Dense(1,activation="sigmoid"))
```

```python
model.compile(loss="binary_crossentropy",optimizer="adam", metrics=["accuracy"])
#classification da sonuç 1 veya 0 olacaksa binary_crossentropy
# 2 den fazla seçenek varsa category_ crossentropy
```

```python
model.fit(x,y, epochs=150, batch_size=10, verbose=1) 
# epoc nöronlar arası kaç defa gelip gider
# batch_size ise veriyi kaçar kaçar işleyecek
# verbose aşağıda yazılan çıkması için olur
```

```python
model.summary() #özeti veriyor
```

```python
scores=model.evaluate(x,y)
# değerlendirir sonuçu verir
```

```python
print("%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))
```

```python
history=model.fit(x,y, validation_split=0.20, epochs=200, batch_size=10, verbose=2)
# history adlı değişkene ataadık verileri kaydettik bunu kullanabiliriz
```

## Derin Öğrenme (Regression)

```python
import pandas as pd
df=pd.read_csv("kc_house.csv")
```

```python
df
```

```python
x=df.drop("price",axis=1)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
```

```python
x=scaler.fit_transform(x)
```

```python
y=df["price"]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import  r2_score, mean_squared_error, mean_absolute_error
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.20, random_state=42)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation
from tensorflow.keras.optimizers import Adam
```

```python
x.shape
```

```python
model=Sequential()
model.add(Dense(81, activation='relu'))
model.add(Dense(40, activation='relu'))
model.add(Dense(27, activation='relu'))
model.add(Dense(19, activation='relu'))
model.add(Dense(5, activation='relu'))
model.add(Dense(3, activation='relu'))
model.add(Dense(1))
model.compile(optimizer='Adam', loss='mse')
```

```python
model.fit(x=x_train, y=y_train, validation_data=(x_test, y_test), batch_size=128, epochs=450, verbose=1)
```

```python
pd.DataFrame(x)
```

```python
model.summary()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
loss_df=pd.DataFrame(model.history.history)
loss_df.plot(figsize=(12,8));
```

```python
pred=model.predict(x_test)
```

```python
(mean_squared_error(y_test,pred))**0.5
```

```python
r2_score(y_test,pred)
```

## 60 bin elyazması rakamların sınıflandırılması

```python
import tensorflow as tf
from tensorflow import keras
```

```python
mnist=keras.datasets.mnist
(train_images, train_labels), (test_images, test_labels) = mnist.load_data()
```

```python
train_images.shape
```

```python
train_images,
```

```python
train_labels
```

```python
from matplotlib import pyplot as plt
plt.imshow(train_images[59999], cmap='gray'); # burada 8 i n yazılışı
```

```python
#Normalize
train_images=train_images/255.0
test_images=test_images/255.0
```

```python
#Define model
#CNN - Convolutional Neural Network 
model=keras.Sequential([
                        keras.layers.InputLayer(input_shape=(28,28)),
                        keras.layers.Reshape(target_shape=(28,28,1)),
                        keras.layers.Conv2D(filters=12, kernel_size=(3,3), activation=tf.nn.relu),
                        keras.layers.MaxPooling2D(pool_size=(2,2)),
                        keras.layers.Flatten(),
                        keras.layers.Dense(10)
])
# CNN convolutional neural networks
```

```python
model.compile(optimizer='adam',
              loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
              metrics=["accuracy"])
```

```python
model.fit(train_images, train_labels, validation_split=0.1, epochs=10)
```

```python
_, accuracy=model.evaluate(test_images, test_labels)
print('Accuracy: %.2f' % (accuracy*100))
```

```python
class_names= ["T-shirt/top", "Trouser", "Pullover", "Dress", "Coat", "Sandal", "Shirt", "Sneaker", "Bag", "Ankle boot"]
```

```python
plt.figure(figsize=(10,10))
for i in range(25):
    plt.subplot(5,5,i+1)
    plt.xticks([])
    plt.yticks([])
    plt.grid(False)
    plt.imshow(train_images[i], cmap=plt.cm.binary)
    plt.xlabel(class_names[train_labels[i]])
plt.show();
```

```python
model=tf.keras.Sequential([
                           tf.keras.layers.Flatten(input_shape=(28, 28)),
                           tf.keras.layers.Dense(128, activation="relu"),
                           tf.keras.layers.Dense(128, activation="relu"),
                           tf.keras.layers.Dense(10)
])
```

```python
model.compile(optimizer="adam",
              loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
              metrics=["accuracy"])
```

```python
model.fit(train_images, train_labels, epochs=10, verbose=1)
```

```python
test_loss, test_acc = model.evaluate(test_images, test_labels, verbose=2)
print("\nTest accuracy:", test_acc)
```

```python
model_json=model.to_json()
with open("model_minst.json", "w") as json_file:
    json_file.write(model_json)

model.save_weights("model_mnist.h5")
print("Saved model to disk")
```
