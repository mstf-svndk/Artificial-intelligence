```python
from sklearn.decomposition import PCA
```

```python
import pandas as pd
```

```python
df=pd.read_csv("PastHires.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
# df['Level of Education']=df['Level of Education'].map({'BS':1,'MS':2, 'PhD':3})
# df.replace({"Y":1, "N":0})
#  verideki y ne n değerlerini 1 ve 0 ile değiştirdik
```

```python
df.replace({"Y":1,"N":0,"BS":1,"MS":2,"PhD":3},inplace=True)
# verideki y ne n değerlerini 1 ve 0 ile değiştirdik
```

```python
df.sample(10)
```

```python
df[df["Employed?"]==1]
```

```python
y=df.Hired
# işe alıp almadığımızın sonucunu bu veride tutuyoruz
```

```python
x=df.drop("Hired",axis=1)
```

```python
from sklearn.tree import DecisionTreeClassifier
clf=DecisionTreeClassifier()
clf.fit(x,y)
```

```python
clf.predict([[10,1,4,3,1,0]])
# çıktı 1 ise işe almışız array 1 çıkmış
```

```python
from sklearn.decomposition import PCA
```

```python
pca=PCA(n_components=2)
X_r=pca.fit_transform(x)
# 6 satırı 2 satıra indirdik
```

```python
X_r
#x ile X_r aynı veriyi temsil ediyor
```

```python
pca.explained_variance_ratio_
#veri kaybolmuşmu buna bakıyoruz
```

```python
sum(pca.explained_variance_ratio_)
#verinin yüzde 97 si duruyor
```

* 6 sütunu 2 sütuna indiriyoruz bir sütunun diğer sütun üzerine izdüşümünü alıyoruz
